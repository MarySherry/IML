A histogram can show the presence of outliers and distribution of data

```python
features = list(set(countries.columns) - set(['country', 'region', 'europe']))

countries[features].hist(figsize=(20,12));
```

Outliers may be plotted as individual points in boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(nrows=3, ncols=6, figsize=(30, 20))

for idx, feat in  enumerate(features):
    sns.boxplot(x='europe', y=feat, data=countries, ax=axes[int(idx / 6), idx % 6])
    axes[int(idx / 6), idx % 6].set_xlabel('europe')
    axes[int(idx / 6), idx % 6].set_ylabel(feat);
```

The interquartile range (IQR) is a measure of statistical dispersion, being equal to the difference between 75th and 25th percentiles. Box plot use the IQR method to display data and outliers. The interquartile range can be used to identify outliers in data regardless of the distribution.

```python
Q1 = countries[features].quantile(0.25)
Q3 = countries[features].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

The data point where we have False that means these values are valid whereas True indicates presence of an outlier.

```python
(countries[features].head(10) < (Q1 - 1.5 * IQR)) |(countries[features].head(10) > (Q3 + 1.5 * IQR))
```

Outliers shouldn't be dropped (too much cases) but at the same time we might want to correct them

```python
countries_out = countries[features][~((countries[features] < (Q1 - 1.5 * IQR)) |(countries[features] > (Q3 + 1.5 * IQR))).any(axis=1)]
print(countries[features].shape)
print(countries_out.shape)
```

```python
outliers = countries[features][(countries[features] < (Q1 - 1.5 * IQR)) |(countries[features] > (Q3 + 1.5 * IQR))].count()
outliers.sort_values(axis=0, ascending=False, inplace=True)
outliers.head(18)
```

AdaBoost can be sensitive to outliers / label noise because it is fitting a classification model (an additive model) to an exponential loss function, and the exponential loss function is sensitive to outliers/label noise. AdaBoost increase the weights of misclassified points on every iteration and therefore might put high weights on these outliers as they tend to be often misclassified.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import StandardScaler

y = countries['europe']
clf = AdaBoostClassifier(n_estimators=50, algorithm = 'SAMME')

X_train, X_test, y_train, y_test = train_test_split(countries[features], y.values, test_size=0.20) 
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)

print ("AdaBoost: ", accuracy)
```

Filter outliers using adaboost. Objects with high weights could be removed as outliers.

```python
weights = clf.estimator_weights_
clfs = clf.estimators_
```

```python
from math import exp
import numpy as np

a = []
alpha = 1/len(X_train)
for i in range(len(X_train)):
    a.append(alpha)
for j in range(len(clfs)):
    clfs[j].fit(X_train, y_train)
    pred_train = clfs[j].predict(X_train)
    for i in range(len(X_train)):
        if y_train[i] == pred_train[i]:
            a[i] = a[i] * exp(-weights[j])
        else:
            a[i] = a[i] * exp(weights[j])

a = np.array(a)
ind = np.where(a > 1)[0]
ind = ind.tolist()
```

```python
X = countries[features].drop(ind)
y = y.drop(ind)
```

Retrain the model without outliers

```python
clf.fit(X, y)
accuracy = clf.score(X_test, y_test)

print ("AdaBoost: ", accuracy)
```
